```python
import numpy as np
import pandas as pd
import random as rand
```

```python
#reading data set
df=pd.read_csv('rank.csv')
df.head()
df.size
# Creating a new column filled with '1'
ones = pd.Series([1] * 85, name='ones')
# Concatenating the new column to the existing DataFrame
updatedDataSet = pd.concat([df, ones], axis=1)
# Save the updated DataFrame to a new CSV file
updatedDataSet.to_csv('updatedDataSet.csv', index=False)
df=pd.read_csv('updatedDataSet.csv')
df.head(84)
```

```python
x0 = df['ones']
x1 = df['SAT']
x2 =df['GPA']
y=df['Rand 1,2,3']
df.info()
```

```python
#parameter initialization :
theta =[0,0,0]
```

```python
# cost function 
J=0
finalCost = []
# count = 0
for i in range(100):
    #calculating hypothesis function:
    m= len(y)
    h = []
    #for each row we are calculating hypothesis:
    for j in range(m):
        h.append(theta[0]*x0[j]+theta[1]*x1[j]+theta[2]*x2[j])
        # print(h[j])
    # print(count," hypothesis value : ",h)
    # count=count+1
    #calulating cost function :
    error =0
    for j in range(m):
        error =error+((h[j]-y[j])**2)
    J=(1/2*m)*error
    finalCost.append(J)
    # print(J)

    #calculating gradient descent:
    sum0=0
    sum1=0
    sum2=0
    for k in range(m):
        sum0=sum0+(h[k])*x0[k]
        sum1=sum1+(h[k])*x1[k]
        sum2=sum2+(h[k])*x2[k]

    #initializing learning rate : 
    alpha=0.01
    theta[0]=(1/m)*(theta[0]-alpha*sum0)
    theta[1]=(1/m)*(theta[1]-alpha*sum1)
    theta[2]=(1/m)*(theta[2]-alpha*sum2)

    print(theta)
```
