```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

```

```python
# Generate random data points with two distinct clusters
data, labels = make_blobs(n_samples=10000, centers=4, random_state=50, cluster_std=10)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Random Data with 4 Clusters')
plt.show()
```

```python
# Specify the number of clusters (K)
k_value = 4

# Create the KMeans object and fit the data
kmeans = KMeans(n_clusters=k_value)
kmeans.fit(data)

# Get the centroids and labels from K-means
centroids = kmeans.cluster_centers_
predicted_labels = kmeans.labels_
```

```python
# Plot the clustered data points and centroids
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='rainbow')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=150, c='black', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.legend()
plt.show()
```
